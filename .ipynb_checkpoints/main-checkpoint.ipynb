### Imports

```python
#Allow import from parent folder
import os, sys
dir2 = os.path.abspath('')
dir1 = os.path.dirname(dir2)
if not dir1 in sys.path: sys.path.append(dir1)

#Needed imports for script
from utils.nii_utils import *
from utils.fitting_utils import *
from DNCNN.network_dncnn import DnCNN as net

import numpy as np
import math
import matplotlib.pyplot as plt
import cv2
import torch

from skimage.restoration import denoise_nl_means, estimate_sigma
```

## Constants

```python
#mean filter
MeanKernelSize = (3, 3) 
#NLM filter
patch_kw = dict(patch_size=3,      # 5x5 patches
                patch_distance=6,  # 13x13 search area
                channel_axis=-1)
```

### Local function definitions

```python
def get_GT(param_folder, file_prefix):
    S0 = load_nii(os.path.join(param_folder, file_prefix + "S0.nii"))
    # S0 = ndimage.zoom(S0, [3, 3, 1], order=1)
    MD = load_nii(os.path.join(param_folder, file_prefix + "MD.nii"))
    # MD = ndimage.zoom(MD, [3, 3, 1], order=1)

    return S0, MD

def load_masks(filename="masks.npy",downscale=False):
    masks = np.load(filename)
    if downscale:
        masks = ndimage.zoom(masks, [1/3, 1/3, 1], order=1)
    return masks

def simulate_y(S0,MD,b,noise_type="normal", noise_sigma=0):
    print("Generating data with "+noise_type+" noise, sigma="+str(noise_sigma))

    y = np.zeros((np.shape(S0)[0], np.shape(S0)[1], np.shape(S0)[2], len(b)))
    for slice_no in range(np.shape(S0)[2]):
        for bi_idx, bi in enumerate(b):
            clean_sim_y = S0[:, :, slice_no] * np.power(math.e, -bi * MD[:, :, slice_no])
            match noise_type:
                case "normal":
                    gen_noise = np.random.normal(0, noise_sigma, np.shape(S0[:, :, slice_no]))
                    y[:, :, slice_no, bi_idx] = clean_sim_y + gen_noise
                case "rician":
                    gen_noise = noise_sigma*(np.random.normal(0, 1, np.shape(S0[:, :, slice_no]))+np.random.normal(0, 1, np.shape(S0[:, :, slice_no]))*1j)
                    y[:, :, slice_no, bi_idx] = np.abs(clean_sim_y+gen_noise)

            # if np.mean(gen_noise ** 2) == 0:
            #     print("No noise PSNR- inf")
            # else:
            #     print("Generated noise at PSNR of ", 20 * math.log10(1500 / math.sqrt(np.mean(gen_noise ** 2))), "DB")

    return y


def filter_y(y, filter_type):
    match filter_type:
        case 'none':
            filtered_y = y.copy()
        case 'mean':
            filtered_y = cv2.blur(y, MeanKernelSize)
        case 'nlm':
            noise_img  = y
            sigma_est  = np.mean(estimate_sigma(noise_img, channel_axis=-1))
            filtered_y = denoise_nl_means(noise_img, 
                                          h=1.15 * sigma_est,
                                          fast_mode=False,
                                          multichannel=False,
                                          **patch_kw)
        case 'dncnn-blind':
            noise_img = y
            filtered_y = filter_dncnn(noise_img, 'dncnn_gray_blind', 20) #nb=20 for blind, 17 for other  
            
    return filtered_y


def filter_parami(parami, filter_type):
    match filter_type:
        case 'none':
            filtered_parami = parami.copy()
        case 'mean':
            filtered_parami = cv2.blur(parami, MeanKernelSize)
        case 'nlm':
            noise_img  = parami
            sigma_est  = np.mean(estimate_sigma(noise_img, channel_axis=-1))
            filtered_parami = denoise_nl_means(noise_img, 
                                          h=1.15 * sigma_est,
                                          fast_mode=False,
                                          multichannel=False,
                                          **patch_kw)
        case 'dncnn-blind':
            noise_img = parami
            filtered_parami = filter_dncnn(noise_img, 'dncnn_gray_blind', 20) #nb=20 for blind, 17 for other  
            
    return filtered_parami


def plot_y(y):
    slice_no = np.shape(y)[2]
    for i in range(slice_no):
        plt.subplot((slice_no+1) // 3 + 1, 3, i+1)
        plt.imshow(y[:, :, i, 0], cmap='gray')
    plt.gcf().set_size_inches(40, 80)
    plt.gcf().tight_layout(pad=0)
    plt.show()

def plot_param(param):
    slice_no = np.shape(param)[2]
    for i in range(slice_no):
        plt.subplot((slice_no+1) // 3 + 1, 3, i+1)
        plt.imshow(param[:, :, i], cmap='gray')
    plt.gcf().set_size_inches(40, 80)
    plt.gcf().tight_layout(pad=0)
    plt.show()
    
def calc_RMSE(param, param_, masks):
    paramcopy = param.copy()
    paramcopy[np.invert(masks)] = np.nan
    param_copy = param_.copy()
    param_copy[np.invert(masks)] = np.nan
    flatparam = paramcopy.flatten()
    flatparam_ = param_copy.flatten()

    flatparam = np.delete(flatparam, np.where(np.isnan(flatparam_)))
    flatparam_ = np.delete(flatparam_, np.where(np.isnan(flatparam_)))

    param_RMSE = math.sqrt(np.square(np.subtract(flatparam, flatparam_)).mean())

    return param_RMSE, np.mean(flatparam)

def filter_dncnn(img, model,nb):
    device = torch.device('cpu')#'cuda' if torch.cuda.is_available() else 'cpu'

    model_path = os.path.join('../DNCNN/models', model + '.pth')
    model = net(in_nc=1, out_nc=1, nc=64, nb=nb, act_mode='R')
    #model = net(in_nc=n_channels, out_nc=n_channels, nc=64, nb=nb, act_mode='BR')  # use this if BN is not merged by utils_bnorm.merge_bn(model)
    model.load_state_dict(torch.load(model_path), strict=True)
    model.eval()
    for k, v in model.named_parameters():
        v.requires_grad = False

    model = model.to(device)

    img_min = np.min(img)
    img_max = np.max(img)

    img_scale = (img-img_min)/(img_max-img_min)
    img_L = np.expand_dims(img_scale, axis=2)
    img_L = torch.from_numpy(np.ascontiguousarray(img_L)).permute(2, 0, 1).float().unsqueeze(0)
    img_L = img_L.to(device)
    img_E = model(img_L)

    filtered_img = img_E.cpu().detach().numpy()[0, 0, :, :]
    filtered_img = filtered_img*(img_max-img_min)+img_min
    return filtered_img

def replaceNANinpar(p,n=1):
    shape_p = np.shape(p)
    for slice_no in range(shape_p[2]):
        for i in range(shape_p[0]):
            for j in range(shape_p[1]):
                for pi in range(shape_p[3]):
                    if np.isnan(p[i, j, slice_no,pi]):
                        rmin = max(i - n, 0)
                        rmax = min(i + n, shape_p[1]-1)
                        cmin = max(j - n, 0)
                        cmax = min(j + n, shape_p[0]-1)
                        
                        print(np.all(np.isnan(p[rmin:rmax, cmin:cmax, slice_no,pi])))
                        if np.all(np.isnan(p[rmin:rmax, cmin:cmax, slice_no,pi])):
                            p[i, j, slice_no,pi] = 0
                        else:
                            p[i, j, slice_no,pi] = np.nanmean(p[rmin:rmax, cmin:cmax, slice_no,pi])
                            
                            
        return pi
```

### Generate noisy data

```python
b = [0.1, 0.3, 0.5, 0.8]
y = simulate_y(S0, MD,b, noise_type="rician", noise_sigma=10)
```

```python
print("Simulated noisy plots for b="+str(b[0]))
plot_y(y)
```

## Filter slices 

```python
filter_type = "nlm"

filtered_y = np.zeros(np.shape(y))
for slice_no in range(np.shape(y)[2]):
    for bindx in range(np.shape(y)[3]):
        filtered_y[:, :, slice_no, bindx] = filter_y(y[:, :, slice_no, bindx],filter_type)

print("Filetered slices")
plot_y(filtered_y)
```

## Fit parameters

```python
p = fit_parameters(b, filtered_y, masks)
backup_p = p.copy()
```

```python
p = backup_p.copy()
```

```python
S0_ = p[:,:,:,0]
MD_ = p[:,:,:,1]

MD_RMSE, mean_MD = calc_RMSE(MD, MD_, masks)
print("MD RMSE = " + str(np.round(MD_RMSE,decimals=2)) + 
" mean = " + str(np.round(mean_MD,decimals=2)) + 
" RMSE/Mean = " + str(np.round(100*MD_RMSE/mean_MD,decimals=1)) + "%")
print("Masked fitted MD")

plot_param(MD_)
```

## ADMM cycle

```python
admm_lambda = 20 #20
filter_type = "dncnn-blind"

for iter_id in range(10):
    filtered_MD = np.zeros(np.shape(MD))
    replaceNANinpar(p,n=1)
    S0_ = p[:,:,:,0]
    MD_ = p[:,:,:,1]

    MD_RMSE, mean_MD = calc_RMSE(MD, MD_, masks)
    print("MD RMSE = " + str(np.round(MD_RMSE,decimals=2)) + 
  " mean = " + str(np.round(mean_MD,decimals=2)) + 
  " RMSE/Mean = " + str(np.round(100*MD_RMSE/mean_MD,decimals=1)) + "%")
    print("Masked fitted fitlered MD")
    for slice_no in range(np.shape(filtered_MD)[2]):
        filtered_MD[:, :, slice_no] = filter_parami(MD_[:, :, slice_no],filter_type)  
    plot_param(filtered_MD)
    p = fit_parameters(b, filtered_y, masks, filtered_MD, mu=admm_lambda*iter_id) 
```

```python
S0_ = p[:,:,:,0]
MD_ = p[:,:,:,1]
```
